Importing Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Importing Dataset

```python
url = 'https://raw.githubusercontent.com/niyathimariya/advertising-sales-prediction/main/advertising.csv'
ad = pd.read_csv(url)
ad.head(10)
```

Data Preprocessing

Checking for null values

```python
print(ad.isnull().sum())
```

Perform linear regression for TV

```python
X_tv = ad[['TV']]
y = ad['Sales']
lm_tv = LinearRegression().fit(X_tv, y)
```

Summary for TV regression

```python
print("\nTV:")
print("Intercept:", lm_tv.intercept_)
print("Coefficient:", lm_tv.coef_)
```

Scatter plot for TV

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=ad, x='TV', y='Sales', alpha=0.2)
plt.plot(ad['TV'], lm_tv.predict(X_tv), color='red')
plt.title('TV')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

Perform linear regression for Radio

```python
X_radio = ad[['Radio']]
lm_radio = LinearRegression().fit(X_radio, y)
```

Summary for Radio regression

```python
print("\nRadio:")
print("Intercept:", lm_radio.intercept_)
print("Coefficient:", lm_radio.coef_)
```

Scatter plot for Radio

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=ad, x='Radio', y='Sales', alpha=0.2)
plt.plot(ad['Radio'], lm_radio.predict(X_radio), color='red')
plt.title('Radio')
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.show()
```

Perform linear regression for Newspaper

```python
X_newspaper = ad[['Newspaper']]
lm_newspaper = LinearRegression().fit(X_newspaper, y)
```

Summary for Newspaper regression

```python
print("\nNewspaper:")
print("Intercept:", lm_newspaper.intercept_)
print("Coefficient:", lm_newspaper.coef_)
```

Scatter plot for Newspaper

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=ad, x='Newspaper', y='Sales', alpha=0.2)
plt.plot(ad['Newspaper'], lm_newspaper.predict(X_newspaper), color='red')
plt.title('Newspaper')
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.show()
```

Perform multiple linear regression for TV + Radio

```python
X_tv_radio = ad[['TV', 'Radio']]
lm_tv_radio = LinearRegression().fit(X_tv_radio, y)
```

Summary for TV + Radio regression

```python
print("\nTV + Radio:")
print("Intercept:", lm_tv_radio.intercept_)
print("Coefficients:", lm_tv_radio.coef_)
```

Scatter plot for TV + Radio

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=ad, x='TV', y='Sales', alpha=0.2)
sns.scatterplot(data=ad, x='Radio', y='Sales', alpha=0.2)
plt.plot(ad['TV'], lm_tv_radio.predict(X_tv_radio), color='red')
plt.title('TV + Radio')
plt.xlabel('TV and Radio')
plt.ylabel('Sales')
plt.legend(['Regression Line', 'TV', 'Radio'])
plt.show()
```

Perform multiple linear regression for TV + Newspaper

```python
X_tv_newspaper = ad[['TV', 'Newspaper']]
lm_tv_newspaper = LinearRegression().fit(X_tv_newspaper, y)
```

Summary for TV + Newspaper regression

```python
print("\nTV + Newspaper:")
print("Intercept:", lm_tv_newspaper.intercept_)
print("Coefficients:", lm_tv_newspaper.coef_)
```

Scatter plot for TV + Newspaper

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=ad, x='TV', y='Sales', alpha=0.2)
sns.scatterplot(data=ad, x='Newspaper', y='Sales', alpha=0.2)
plt.plot(ad['TV'], lm_tv_newspaper.predict(X_tv_newspaper), color='red')
plt.title('TV + Newspaper')
plt.xlabel('TV and Newspaper')
plt.ylabel('Sales')
plt.legend(['Regression Line', 'TV', 'Newspaper'])
plt.show()
```

Perform multiple linear regression for TV + Radio + Newspaper

```python
X_all = ad[['TV', 'Radio', 'Newspaper']]
lm_all = LinearRegression().fit(X_all, y)
```

 Summary for TV + Radio + Newspaper regression

```python
print("\nTV + Radio + Newspaper:")
print("Intercept:", lm_all.intercept_)
print("Coefficients:", lm_all.coef_)
```

Scatter plot for TV + Radio + Newspaper

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=ad, x='TV', y='Sales', alpha=0.2)
sns.scatterplot(data=ad, x='Radio', y='Sales', alpha=0.2)
sns.scatterplot(data=ad, x='Newspaper', y='Sales', alpha=0.2)
plt.plot(ad['TV'], lm_all.predict(X_all), color='red')
plt.title('TV + Radio + Newspaper')
plt.xlabel('TV, Radio, and Newspaper')
plt.ylabel('Sales')
plt.legend(['Regression Line', 'TV', 'Radio', 'Newspaper'])
plt.show()
```
